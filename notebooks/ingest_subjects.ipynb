```python
def date_from_string(date_str):
    """return a datetime.date object from a string in the following format: YYYY-MM-DD

    Args:
        date_str (str): date string to be converted

    Returns:
        datetime.date: string converted to a datetime.date object
    """
    return datetime.strptime(date_str, "%Y-%m-%d").date()
```
